# 1. Lectura de datos

```
# Nombre de la base de datos en Databricks
database_name = "default"
# Nombre de la tabla que deseas leer
table_name = "gpudata_csv"

# Leer la tabla en un DataFrame de Spark
df = spark.table(f"{database_name}.{table_name}")

# Mostrar las primeras filas de la tabla
display(df)
```

# 2. Perfilamiento básico de datos

```
from pyspark.sql.functions import col, mean, stddev, min, max, countDistinct, approxCountDistinct
from pyspark.sql.types import IntegerType, LongType, DoubleType

# Mostrar información general del DataFrame
print("Información general del DataFrame:")
print(f"Número total de filas: {df.count()}")
print("Número de columnas:", len(df.columns))

# Obtener información particular de cada columna
print("\nInformación de cada columna:")
for col_name in df.columns:
    print(f"\nColumna: {col_name}")
    print(f"Tipo de datos: {df.schema[col_name].dataType}")
    print(f"Número de valores no nulos: {df.filter(col(col_name).isNotNull()).count()}")
    print(f"Número de valores nulos: {df.filter(col(col_name).isNull()).count()}")
    
    # Si la columna es numérica, calcular estadísticas descriptivas
    if isinstance(df.schema[col_name].dataType, (IntegerType, LongType, DoubleType)):
        stats = df.select(mean(col_name).alias('Mean'),
                          stddev(col_name).alias('Stddev'),
                          min(col_name).alias('Min'),
                          max(col_name).alias('Max')).collect()[0]
        print(f"Media: {stats['Mean']}, Desviación estándar: {stats['Stddev']}")
        print(f"Mínimo: {stats['Min']}, Máximo: {stats['Max']}")
        
        # Histograma
        df.select(col_name).rdd.flatMap(lambda x: x).histogram(10)
        
    # Si la columna es categórica, mostrar valores únicos y frecuencia
    elif df.schema[col_name].dataType in ['StringType']:
        unique_values = df.groupBy(col_name).count().orderBy('count', ascending=False).collect()
        print("Valores únicos:")
        for val in unique_values:
            print(f"{val[col_name]}: {val['count']}")
        
    # Mostrar algunos ejemplos de valores (opcional)
    example_values = df.select(col_name).sample(False, 0.1).limit(5).collect()
    print("Ejemplos de valores:")
    for row in example_values:
        print(row[col_name])
```

```
from pyspark.sql.functions import avg, desc
import matplotlib.pyplot as plt

# Calcular el promedio del precio y la potencia para cada serie o modelo de gráfica
avg_price_power = df.groupBy('Name').agg(avg('Price').alias('AvgPrice'), avg('Boost Clock(MHz)').alias('AvgPower'))

# Seleccionar los 8 modelos con los precios y la potencia promedio más altos
top_avg_price_power = avg_price_power.orderBy(desc('AvgPrice'), desc('AvgPower')).limit(8)

# Convertir el DataFrame resultante a Pandas para visualizar el gráfico de barras
top_avg_price_power_pandas = top_avg_price_power.toPandas()

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
bars = plt.bar(top_avg_price_power_pandas['Name'], top_avg_price_power_pandas['AvgPower'], label='Potencia promedio', color='orange')

# Colocar el precio en la parte derecha de las barras
for bar, price in zip(bars, top_avg_price_power_pandas['AvgPrice']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f"${price:.2f}", ha='center', va='bottom')

plt.xlabel('Modelo')
plt.ylabel('Potencia Promedio (MHz)')
plt.title('Top 8: Mas potentes calidad-precio')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import col, row_number

# Guardar el estado actual del DataFrame
df_state = df

# Filtrar temporalmente el DataFrame para excluir los productores especificados y filas con datos nulos
df = df.filter(~col('Producer').isin(producers_to_remove)) \
       .filter(col('Price').isNotNull() & col('Boost Clock(MHz)').isNotNull())

# Calcular la métrica de relación calidad-precio
df = df.withColumn('QualityPriceRatio', col('Price') / col('Boost Clock(MHz)'))

# Definir una ventana por productor para ordenar por la métrica de relación calidad-precio
window_spec = Window.partitionBy('Producer').orderBy('QualityPriceRatio')

# Asignar un número de fila para cada producto dentro de su productor, ordenado por la métrica de relación calidad-precio
df = df.withColumn('rank', row_number().over(window_spec))

# Filtrar los productos con el mejor ranking por productor
best_products_by_producer = df.filter(col('rank') == 1).drop('QualityPriceRatio', 'rank')

# Mostrar los mejores productos por productor
best_products_by_producer.show()

```

```
# Guardar el estado actual del DataFrame
df_state = df

# Filtrar temporalmente el DataFrame para excluir los productores especificados y filas con datos nulos
df = df.filter(~col('Producer').isin(producers_to_remove)) \
       .filter(col('Price').isNotNull() & col('Boost Clock(MHz)').isNotNull())

# Calcular la métrica de relación calidad-precio
df = df.withColumn('QualityPriceRatio', col('Price') / col('Boost Clock(MHz)'))

# Definir una ventana por productor para ordenar por la métrica de relación calidad-precio en orden descendente
window_spec = Window.partitionBy('Producer').orderBy(df['QualityPriceRatio'].desc())

# Asignar un número de fila para cada producto dentro de su productor, ordenado por la métrica de relación calidad-precio
df = df.withColumn('rank', row_number().over(window_spec))

# Filtrar los productos con el peor ranking por productor (los que tienen la relación calidad-precio más alta)
worst_products_by_producer = df.filter(col('rank') == 1).drop('QualityPriceRatio', 'rank')

# Mostrar los peores productos por productor
worst_products_by_producer.show()

# Restaurar el DataFrame al estado original
df = df_state
```

```
import matplotlib.pyplot as plt

# Obtener los datos para el gráfico de barras
productores_mejor = best_products_by_producer.select('Producer').rdd.flatMap(lambda x: x).collect()
productores_peor = worst_products_by_producer.select('Producer').rdd.flatMap(lambda x: x).collect()

# Precios de las mejores GPUs calidad-precio
mejor_precios = best_products_by_producer.select('Price').rdd.flatMap(lambda x: x).collect()

# Precios de las peores GPUs calidad-precio
peores_precios = worst_products_by_producer.select('Price').rdd.flatMap(lambda x: x).collect()

# Asegurarse de que ambos conjuntos de datos tengan la misma longitud
if len(productores_mejor) > len(productores_peor):
    productores_peor += [None] * (len(productores_mejor) - len(productores_peor))
    peores_precios += [None] * (len(productores_mejor) - len(peores_precios))
elif len(productores_peor) > len(productores_mejor):
    productores_mejor += [None] * (len(productores_peor) - len(productores_mejor))
    mejor_precios += [None] * (len(productores_peor) - len(mejor_precios))

# Configuración del gráfico de barras
x = range(len(productores_mejor))
ancho_barra = 0.35

# Crear el gráfico de barras
plt.figure(figsize=(12, 8))
plt.bar(x, mejor_precios, width=ancho_barra, label='Mejor Precio', color='green')
plt.bar([i + ancho_barra for i in x], peores_precios, width=ancho_barra, label='Peor Precio', color='red')

# Configuración de ejes y etiquetas
plt.xlabel('Productor')
plt.ylabel('Precio')
plt.title('Mejor y Peor Precio por Productor')
plt.xticks([i + ancho_barra / 2 for i in x], productores_mejor, rotation=45)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
